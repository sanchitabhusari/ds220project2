```python
import pandas as pd

# Load your dataset into a Pandas DataFrame
# Replace 'your_dataset.csv' with the actual file path or URL of your dataset
nobel_prize = pd.read_csv('export.csv')

# The column containing years is named 'year'
if 'year' in nobel_prize.columns:
    # Find the earliest year
    earliest_year = nobel_prize['year'].min()

    print(f"The earliest year in the dataset is: {earliest_year}")
else:
    print("The 'year' column does not exist in the dataset.")
```

```python
import matplotlib.pyplot as plt
import pandas as pd
year_sorted = nobel_prize['year'].value_counts()
year_chart = year_sorted.head(20)

year_chart
```

```python
# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(year_chart, labels= [2001, 2022, 2019, 2009, 2011, 2013, 2014, 2002, 2005, 2000, 2018, 2021, 1996, 1973, 1994, 1995, 1988, 2004, 2008, 1998], autopct='%1.1f%%', startangle=140)

# Add a title
plt.title('Top Ten Years Nobel Prices are Awarded')

# Show the pie chart
plt.show()
```

```python
# The gender column is named 'gender'
gender_column = 'gender'

# Count the number of females and males
gender_counts = nobel_prize[gender_column].value_counts()

# Display the results
print("Number of Females:", gender_counts.get('female', 0))
print("Number of Males:", gender_counts.get('male', 0))
```

```python
plt.bar(gender_counts.index, gender_counts)

# Adding labels and title
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Bar Chart of Gender')
```

```python
# The category column is named 'category'
category_column = 'category'

# Count the occurrences of each item within the category column
category_counts = nobel_prize[category_column].value_counts()

# Display the results
print("Category Counts:")
print(category_counts)
```

```python
plt.bar(category_counts.index, category_counts)

# Adding labels and title
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Bar Chart of Categories')
```

```python
# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(category_counts, labels= ['medicine', 'physics', 'chemistry', 'peace', 'literature', 'economics'], autopct='%1.1f%%', startangle=140)

# Add a title
plt.title('Category Breakdown')

# Show the pie chart
plt.show()
```

```python
# The country column is named 'country'
country_column = 'country'

# Count the occurrences of each item within the country column
country_counts = nobel_prize[country_column].value_counts()

# Display the results
print("Country Counts:")
print(country_counts)
```

```python
country_sorted = nobel_prize['country'].value_counts()
country_chart = country_sorted.head(10)

country_chart

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(country_chart, labels= country_chart.index, autopct='%1.1f%%', startangle=140)

# Add a title
plt.title('Geographical Breakdown')

# Show the pie chart
plt.show()
```

```python
import datetime as dt
# Convert the 'date' column to datetime format
#nobel_prize['year'] = pd.to_datetime(nobel_prize['year'])

# Filter entries for the year 2001
df_2001 = nobel_prize[nobel_prize['year'] == 2001]

# Find the category with the most entries in the year 2001
most_common_category = df_2001['category'].value_counts().idxmax()
# Display the result
print(f"The category with the most entries in 2001 is: {most_common_category}")
```

```python
share_column = 'share'
share_counts = nobel_prize['share'].value_counts()
# Calculate the average number of shares
average_share = nobel_prize[share_column].mean()

# Display the result
print(f"The average number of shares is: {average_share}")
```

```python
plt.bar(share_counts.index, share_counts)

# Adding labels and title
plt.xlabel('Share')
plt.ylabel('Count')
plt.title('Bar Chart of Collaboration')
```
